```python
endpoint='https://data.nashville.gov/resource/fuaa-r5cm.json'
params={
        "$where": "query_count>100",
        "$limit":"50000"
}
response = requests.get(endpoint, params = params)
```

```python
response
```

```python
response.text
```

```python
res = response.json()
res
```

```python
hun_a_month = pd.DataFrame(res)
hun_a_month
```

```python
hun_a_month['query_count']=hun_a_month['query_count'].astype(int)
```
